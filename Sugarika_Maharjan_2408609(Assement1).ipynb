Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# Define a function to identify outliers
def identify_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data < lower_bound) | (data > upper_bound)]

# Identify outliers in Score and GDP per Capita
score_outliers = identify_outliers(south_asia_df['score'])
gdp_outliers = identify_outliers(south_asia_df['Log GDP per capita'])

outlier_countries = south_asia_df[south_asia_df['score'].isin(score_outliers) | south_asia_df['Log GDP per capita'].isin(gdp_outliers)]
print("Outlier countries based on Score and GDP per Capita:")
outlier_countries
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(south_asia_df["Log GDP per capita"], south_asia_df["score"], label="Non-Outliers", color="skyblue")
plt.scatter(outlier_countries["Log GDP per capita"], outlier_countries["score"], label="Outliers", color="red")
plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.title("GDP per capita vs Score(Outliers Highlights)")
plt.legend()
plt.grid(True)
plt.show()
```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python
def identify_outliers(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
  return outliers

outliers_south_asia_score=identify_outliers(south_asia_df,'score')
outliers_south_asia_gdp=identify_outliers(south_asia_df,'Log GDP per capita')

outliers_middle_east_score=identify_outliers(middle_east_df,'score')
outliers_middle_east_gdp=identify_outliers(middle_east_df,'Log GDP per capita')

outliers_south_asia=pd.concat([outliers_south_asia_score,outliers_south_asia_gdp]).drop_duplicates()
outliers_middle_east=pd.concat([outliers_middle_east_score,outliers_middle_east_gdp]).drop_duplicates()

print("Outliiers for South Asia:")
print(outliers_south_asia[['Country name','score','Log GDP per capita']])

print("\nOutliiers for Middle East:")
print(outliers_middle_east[['Country name','score','Log GDP per capita']])
```

```python
#scatter plot of outliers for south asia
plt.figure(figsize=(12,6))
plt.scatter(south_asia_df['Log GDP per capita'],south_asia_df['score'],color='blue',edgecolors='black',label='Non-Outliers')
plt.scatter(outliers_south_asia['Log GDP per capita'],outliers_south_asia['score'],color='red',edgecolors='black',label='Outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('GDP per capita vs Score(Outliers Highlights)')
plt.legend()
plt.grid(True)
plt.show()

#scatter plot for middle east
plt.figure(figsize=(12,6))
plt.scatter(middle_east_df['Log GDP per capita'],middle_east_df['score'],color='green',edgecolor='black',label='Non-Outliers')
plt.scatter(outliers_middle_east['Log GDP per capita'],outliers_middle_east['score'],color='orange',edgecolor='black',label='Outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('Score')
plt.title('GDP per capita vs Score(Outliers Highlights)')
plt.legend()
plt.grid(True)
plt.show()
```
