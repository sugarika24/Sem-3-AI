Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')
df.head()
```

```python
new_df=df[['Name','Pclass','Sex','Age','Fare','Survived']]
new_df.head()
```

```python
mean_fare=subset_class_df['Fare'].mean()
median_fare=subset_class_df['Fare'].median()
max_fare=subset_class_df['Fare'].max()
min_fare=subset_class_df['Fare'].min()
print('Mean Fare:',mean_fare)
print('\nMedian Fare:\n',median_fare)
print('\nMaximun Fare:\n',max_fare)
print('\nMinimun Fare:\n',min_fare)
```

Problem - 3:

The ’Embarked’ column in the Titanic dataset contains categorical data representing the ports of embarka-
tion:

• ’C’ for Cherbourg

• ’Q’ for Queenstown

• ’S’ for Southampton
Task:
1. Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’,
’Embarked Q’, ’Embarked S’).
2. Add these new columns to the original DataFrame.
3. Drop the original ’Embarked’ column.
4. Print the first few rows of the modified DataFrame to verify the changes.

```python
#Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’, ’Embarked Q’, ’Embarked S’).
if all(col in df.columns for col in['S','C','Q']):
  print("Embarked already encoded")
else:
  df_encoded=pd.get_dummies(df['Embarked'],dtype=int)
  df=pd.concat([df,df_encoded],axis='columns')
if all(em in df.columns for em in ['Embarked']):
  df=df.drop('Embarked',axis='columns')
else:
  print('Embarked already droppped')
```

Problem - 4:

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-
ization to show how the survival distributions vary by gender.

```python
mean_survival_rate=df.groupby('Sex')['Survived'].mean()
mean_survival_rate
```

```python
mean_survival_rate.plot(kind='bar',color=['blue','red'])
plt.xlabel('Sex')
plt.ylabel('Mean Survival Rate')
plt.title('Mean Survival rate by Gender')
plt.show()
```

Problem - 6{Optional}:

Show how the survival rates (’Survived’) vary by age group and passenger class (’Pclass’). Break up the
’Age’ column into five quantiles in your DataFrame, and then compare the means of ’Survived’ by class
and age group. Draw a visualization using a any plotting library to represent this graphically.[link text](https://)

```python
df['AgeGroup']=pd.qcut(df['Age'],5)

#calculate the mean survival rate by agegroup and pclasss
survival_by_age=df.groupby(['AgeGroup','Pclass'])['Survived'].mean().unstack()
survival_by_age.plot(kind='bar',stacked=True)
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Age Group and Passenger Class')
plt.show()
```
